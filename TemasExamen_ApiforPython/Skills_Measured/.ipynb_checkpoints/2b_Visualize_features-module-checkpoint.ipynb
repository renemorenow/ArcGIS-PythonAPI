### position_along_line

```python
itemId_rios = "5ad86d68572540698a1d2f40131546c7" #Red hdrica Colombia
flc_rios = gis.content.get(itemId_rios)
fl_rios = flc_rios.layers[0]
sedf = pd.DataFrame.spatial.from_layer(fl_rios)
rivers_map = gis.map("Colombia")
sedf.spatial.plot(map_widget=rivers_map)
rivers_map
```

```python
fs_rios = fl_rios.query(where="name = 'Ro Magdalena' and popupinfo='Ro'", return_all_records=False, result_record_count=1)
print(len(fs_rios.features), "features into", type(fs_rios))
import copy
```

```python
fs_rios_2pal = copy.deepcopy(fs_rios)
for f in fs_rios_2pal.features:
    geom = Geometry(f.geometry)
    f.geometry = geom.position_along_line(0.01, True)
```

```python
fs_rios_2pal.features[0].geometry
```

```python
map_rios2 = gis.map("Colombia")
#map_rios2.add_layer(fs_rios_bk)
map_rios2.add_layer(fs_rios_2pal)
map_rios2.zoom_to_layer(fs_rios_2pal)
map_rios2
```

### segment_along_line

```python
fs_rios_2segment = copy.deepcopy(fs_rios_bk)
for f in fs_rios_2segment.features:
    geom = Geometry(f.geometry)
    res_sal = geom.segment_along_line(start_measure=0.01, end_measure=0.05, use_percentage=False)
    f.geometry = res_sal

map_rios3 = gis.map("Colombia")
#map_rios2.add_layer(fs_rios_bk)
map_rios3.add_layer(fs_rios_2segment)
map_rios3.zoom_to_layer(fs_rios_2segment)
map_rios3
```

# FeatureLayerManager  (managers submodule)
https://developers.arcgis.com/python/api-reference/arcgis.features.managers.html#featurelayermanager

Allows updating the definition (if access permits) of a FeatureLayer. This class is not created by users directly. An instance of this class, called manager, is available as a property of the FeatureLayer object, if the layer can be managed by the user. Users call methods on this manager object to manage the feature layer.

# FeatureCollectionManager  (managers submodule)
https://developers.arcgis.com/python/api-reference/arcgis.features.managers.html#featurelayercollectionmanager

Allows updating the definition (if access permits) of a FeatureLayerCollection. This class is not created by users directly. An instance of this class, called manager, is available as a property of the FeatureLayerCollection object.
