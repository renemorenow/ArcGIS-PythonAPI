#### Automating mapping tasks

```python
from arcgis.gis import GIS
import pandas as pd
from arcgis.features import GeoAccessor, GeoSeriesAccessor

# Define a function that creates a map of the n largest fires in a state
def create_map_large_fires(s, b, n):
    gis = GIS()
    m = gis.map(s)
    m.basemap = b

    sedf = pd.DataFrame.spatial.from_featureclass(r'D:\EsriTraining\APIforPython\VisualPythonAPI\FirePoints.gdb\FirePoints_Full')
    sedf.loc[sedf.State == s, :].nlargest(n, 'Acres').spatial.plot(map_widget=m)

    return m
```

```python
# Define variables and pass them into the custom function
STATE = 'Utah'
BASEMAP = 'national-geographic'
N_LARGEST = 5 # The largest n fires to show on the map (For example, 5 will show the 5 largest fires by acres burned) 

create_map_large_fires(STATE, BASEMAP, N_LARGEST)
```

### Change map properties using automation

#### Create practice web maps

```python
from arcgis.mapping import WebMap
```

```python
maps_to_copy = ['0e71c646b598465cb21da10450b77d13', 'e4e24b4a44aa4e4ca58620f8ee46d7a0' ,'e1c4fbcb634f4d28a972c0c4018c1592']

for webmap in maps_to_copy:
    wm_item = gis.content.get(webmap)
    new_webmap = WebMap(wm_item)
    webmap_properties = {'title': wm_item.title,
                        'snippet': 'A web map to practice automation workflows',
                        'tags':['automation', 'Esri Training']}
    new_webmap.save(webmap_properties, folder='VisualPythonAPI')
```

#### List items in a content folder

```python
me = gis.users.me
webmaps_to_change = me.items(folder='VisualPythonAPI')

for webmap_item in webmaps_to_change:
    print(webmap_item.title)
```

#### Standardize the basemap among multiple maps

```python
for wm_item in webmaps_to_change:
    webmap = WebMap(wm_item)
    # Add code to set the basemap
    webmap.basemap = "hybrid"
    webmap.update()
```

### Apply renderers to an SEDF map visualization

```python
# Create the SEDF object
sedf = pd.DataFrame.spatial.from_featureclass(r'D:\EsriTraining\APIforPython\VisualPythonAPI\FirePoints.gdb\FirePoints_Montana')
```

```python
map1 = gis.map("Montana")
map1
```

```python
# Modify the following line to plot the data to map1
sedf.spatial.plot(map_widget=map1)
```

#### Use the Simple renderer

```python
map2 = gis.map("Montana")
map2
sedf.spatial.plot(map_widget=map2, renderer_type='s', marker_size=5, cmap="Reds_r", cstep=50, line_width=0)
```

#### Use the Heatmap renderer

```python
map3 = gis.map("Montana")
map3
sedf.spatial.plot(map_widget=map3, renderer_type='h', blur_radius=7)
map3
```

#### Use the Unique renderer

```python
map4 = gis.map("Montana")
map4
```

```python

from arcgis.mapping import display_colormaps
display_colormaps()
```

```python
sedf.spatial.plot(map_widget=map4, renderer_type='u', col='Season', marker_size=5, line_width=0, cmap='prism')
```

```python
map4.legend = True
map4
```

#### Use the Class breaks renderer

```python
map5 = gis.map("Montana")
map5.basemap='dark-gray'
map5
```

```python
sedf.spatial.plot(map_widget=map5,
                  renderer_type='c',   # specify the class breaks renderer
                  method='esriClassifyNaturalBreaks',  # classification algorithm
                  class_count=5, # choose the number of classes
                  col='Acres',  # the attribute field (column) to use
                  cmap='hot', # color map to pick colors from for each class
                  marker_size = 5,
                  alpha=0.6,# specify opacity
                  line_width=0)
```

```python
map5.legend = True
map5
```
